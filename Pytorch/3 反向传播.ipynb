```python
import numpy as np 
import torch 
import matplotlib.pyplot as plt
```

```python
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]
w = torch.Tensor([1.0])
w.requires_grad = True
```

```python
def forward(x):
    return x*w
```

```python
def loss(x, y):
    y_pred = forward(x)
    return (y_pred - y) ** 2
```

```python
print("训练之前:", 4, forward(4).item())
```

```python
for epoch in range(100):
    for x, y in zip(x_data, y_data):
        l = loss(x, y)
        l.backward()
        print(x, y, w.grad.item())
        w.data = w.data - 0.01*w.grad.item()
        w.grad.data.zero_()
    print("epoch:", epoch, "loss:", l.item())
```

```python
print("训练之后:", 4, forward(4).item())
```
